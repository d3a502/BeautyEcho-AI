# 用户反馈聚类 Demo

本 notebook 使用样例评论展示用户反馈聚类思路。当前数据为 demo 数据，不代表真实平台全量数据。

```python
import pandas as pd
from pathlib import Path

data_path = Path('../samples/sample_comments.csv')
df = pd.read_csv(data_path)
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

comments = df['comment'].astype(str).tolist()
vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(2, 3))
X = vectorizer.fit_transform(comments)

model = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = model.fit_predict(X)
df
```

```python
for cluster_id, group in df.groupby('cluster'):
    print(f'Cluster {cluster_id}')
    for text in group['comment'].tolist():
        print('-', text)
    print()
```

## 解读方式

- 将评论聚类为肤质/持妆、价格/学生党、色号/成分/信任等方向。
- 聚类结果用于生成内容策略，而不是直接替代人工判断。
- 入围后可将该分析流程与飞书多维表格和飞书 AI 标签字段结合。
